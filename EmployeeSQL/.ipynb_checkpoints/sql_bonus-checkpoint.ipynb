```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
from password import pw
```

```python
engine = create_engine('postgresql+pg8000://bethkaczor:"pw"@localhost/employeeSQL')
```

```python
connection = engine.connect()
```

```python
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
```

```python
salary_data.head()
```

```python
salary_data.plot=salary_data["salary"].hist()
salary_data.plot.set_title("Salary Frequencies")
salary_data.plot.set_xlabel("Salary")
salary_data.plot.set_ylabel("Number of Employees")
```

```python
titles_data = pd.read_sql("SELECT * FROM titles", connection)
titles_data.head()
```

```python
employees_data = pd.read_sql("SELECT * FROM employees", connection)
employees_data.head()
```

```python
merged_df = pd.merge(salary_data, employees_data, on='emp_no')
merged_df
```

```python
new_merged_df = pd.merge(merged_df, titles_data, left_on="emp_title_id", right_on="title_id")
new_merged_df
```

```python
avg_salary_title = new_merged_df.groupby("title")["salary"].mean()
```

```python
avg_salary_title
```

```python
avg_salary_title_df = pd.DataFrame(data = {"Average Salary": avg_salary_title}).reset_index()
avg_salary_title_df
```

```python
x_axis = avg_salary_title_df["title"]
y_axis = avg_salary_title_df["Average Salary"]
plt.bar(x_axis, y_axis)
plt.xticks(rotation=50)
plt.title("Average Salary by Title")
```
